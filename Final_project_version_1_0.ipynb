now that we got a clean data lets do the next thing. lets find out if a higher education helps to get the loan or not. lets find out first how many were given and how many not

```python

df['loan_status'].value_counts()
```

now a lot of aproven loans huh? well we can see what education has the most aproval loans

```python
# group the education

df.groupby('person_education')['loan_status'].value_counts()

# now we see how many different diplomas give loans next we grafh it

```

we see that the mayordy of loans were given to people with just high school as their higher leven education, we ca nse a visual donw here below.

```python
# Create a bar plot to visualize the loan status by education level
loan_base_on_education = df.groupby('person_education')['loan_status'].value_counts().unstack()


loan_base_on_education.plot(kind='bar', stacked=True)
plt.title('Loan by education Level')
plt.xlabel('level of education')
plt.ylabel('number of loans given')

plt.show()
```

we that there was not a lot of loans aprove in the data, but the eduation that has the most it was high school.
